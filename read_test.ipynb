```python
def get_issue_label(issue_num):
    """
    input:issue item:item
    output:issue item label:list
    """
    label_name = []
    label_describe = []
    for i in issue_num["labels"]:
        label_name.append(i["name"])
        label_describe.append(i["description"])
    
    return label_name,label_describe
```

```python
def get_issue_df(data):
    df = pd.DataFrame(columns=["id","body","label_name","label_describe","comments_url","title"])
    for item in data["issue"]:
        label_name,label_describe = get_issue_label(item)
        pd_data = {"id":item["id"],"body":item["body"],"label_name":label_name,"label_describe":label_describe,"comments_url":item["comments_url"],"title":item["title"]}
        df = df.append(pd_data,ignore_index=True)
    return df
df = get_issue_df(data)
```

```python
import os
def make_raw_data():
    path_list = os.listdir("./issue")
    df = pd.DataFrame(columns=["id","body","label_name","label_describe","comments_url","html_url","title"])
    id_list = set()
    for i in path_list:
        if i.endswith("json"):
            path = os.path.join("./issue/",i)
            data = pd.read_json(path)
            for item in data["issue"]:
                if item["id"] in id_list:
                    continue
                id_list.add(item["id"])
                label_name,label_describe = get_issue_label(item)
                pd_data = {"id":item["id"],"body":item["body"],"label_name":label_name,\
                           "label_describe":label_describe,"comments_url":item["comments_url"],\
                           "html_url":item["html_url"],"title":item["title"]}
                df = df.append(pd_data,ignore_index=True)
            print("append one")
    return df
```

```python
def get_key_word():
    lines = open("./issue分类相关/vul_keywords.txt","r",encoding = "utf-8")
    vul = [i for i in lines]
    vul_keyword = []
    vul_type = []
    for i in vul[0].split(","):
        vul_keyword.append(re.sub("[\s]*","",i).strip("'"))
    for j in vul[1].split(","):
        vul_type.append(re.sub("[\s]*","",j).strip("'"))
    lines.close()
    return vul_keyword,vul_type
```
